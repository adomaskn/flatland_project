```python
import numpy as np
from collections import Counter
from tensorflow import keras
```

```python
data = np.load('flatland_train.npz')
X = data['X']
y = data['y']

y[y != 0] -= 2
X = X / 255.
```

Penkta kodo eilutė pakoregavo kampainių numeraciją. Šeštoji pakeičia reikšmes į paprastesnes.

```python
Counter(y)
```

**Dabar pradesime daryti modeli**

Matome pilnu duomenu matmenis, taciau reikes pasidaryti splita, kad galetumem tureti tam tikra dali duomenu treniravimui, o kita dali - testavimui.

```python
X.shape
```

```python
mask = np.random.random(len(X)) > 0.8
```

Uzmaskuosime train ir test objektus

```python
X_train = X[~mask]
y_train = y[~mask]
X_test = X[mask]
y_test = y[mask]
```

Uzsirasome pirmaji modeli, galinis modelio pridejimas turi buti softmax, nes norime kazko panasesnio i tikimybes

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=(50,50)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(20, activation='relu'))
model.add(keras.layers.Dense(5, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='SGD', metrics='accuracy')
model.summary()
```

```python
model.fit(X_train,y_train, epochs=10)
```

Dabar laikas issisaugoti pirmaji modeli

```python
model.save('model.h5')
```

**Pabandysiu sukurti gal kiek mazesni ir kiek geresni modeli**

```python
model = keras.models.Sequential()

modelnr = 17
best_hist = 0.95
i = 5
while i < 50:
  j = i
  while j > 4:   
    k = j
    while k > 4:     
      m = 100
      while m > 12:
        n = m - 1
        while n > 11:
          l = n - 1
          while l > 10:
            daug = 53 - i - j - k
            #parametru skaiciaus suma, nuo jos priklauso dydis failo ir ji bus ribojimas vienas
            param_sk = (i*i*50 + 50) + (j*j*50*50 + 50) + (k*k*50*50 + 50) + (daug*daug*50*m + m) + (m*n + n) + (n*l + l)
            #tam tikru salygu pridejimas
            if param_sk < 500000:
              if (i + j + k) > 52:
                print('epic fail su ' + str(i) + ' ' + str(j) + ' ' + str(k)+ ' ' + str(m) + ' ' + str(n) + ' ' + str(l))
              else:
                try:
                  model.add(keras.layers.Conv2D(50, kernel_size=(i, i), activation='relu', input_shape=[50, 50, 1]))
                  model.add(keras.layers.Conv2D(50, kernel_size=(j, j), activation='relu'))
                  model.add(keras.layers.Conv2D(50, kernel_size=(k, k), activation='relu'))
                  model.add(keras.layers.Flatten())
                  model.add(keras.layers.Dense(m, activation="relu"))
                  model.add(keras.layers.Dense(n, activation="relu"))
                  model.add(keras.layers.Dense(l, activation="softmax"))
                  model.compile(loss="sparse_categorical_crossentropy", optimizer="adamax", metrics=["accuracy"])
                  hist = model.fit(X_train,y_train, epochs=6)
                  tikslumas = hist.history['accuracy'][-1]
                  if float(tikslumas) > float(best_hist):
                    best_hist = tikslumas
                    modelio_vardas = 'model' + str(modelnr) + '.h5'
                    model.save(modelio_vardas)
                    modelnr = modelnr + 1
                  else:
                    print('nepakankamai didelis tikslumas ' + str(i) + ' ' + str(j) + ' ' + str(k)+ ' ' + str(m) + ' ' + str(n) + ' ' + str(l))
                except ValueError as ve:
                  print('blogi skaiciukai')
            else:
              print('per dideli parametrai su ' + str(i) + ' ' + str(j) + ' ' + str(k) + ' ' + str(m) + ' ' + str(n) + ' ' + str(l))
              


            
            l = l - 5
          n = n - 5
        m = m - 5
      k = k - 5
    j = j - 5
  i = i + 5
```
